oops! That first row is not a data row. Let's get rid of it.

Determine which row you want to drop by its index number. In this case, it is `[0]`. 

```python
df.drop([0])
```

or... we can re-import the data skipping the second row. This is better. Why?

```python
df = pd.read_csv('data/social_explorer.csv',skiprows=[1])
```

```python
df.head()
```

Waaaaaiiit... what happened to the "0" in the FIPS code?

That's right, python interprets this as a integer columns, and crops the leading zeroes. Let's fix this by re-re-importing the data, specifying the FIPS column as a string:

```python
df = pd.read_csv('data/social_explorer.csv',
                 skiprows=[1],
                dtype={'FIPS':str})
```
